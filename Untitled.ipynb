#### Let's try the most basic segmentation ideas.  I create three new columns: weekend/weekday, time of trip (sec from midnight), and age of the user.

```python
def get_weekday(x):
    time = datetime.datetime.strptime(x,'%Y-%m-%d %H:%M:%S.%f')
    
    if time.weekday() in range(5):
        return True
    else:
        return False

def sec_from_mid(x):
    time = datetime.datetime.strptime(x.split()[1],'%H:%M:%S.%f')
    return 3600 * time.hour + 60 * time.minute + time.second

def get_age(x):
    age = 2017 - x
    if age > 500:
        return 'NaN'
    else:
        return age

read_0['weekday'] = read_0['start_time'].apply(get_weekday)
read_0['sec_from_mid'] = read_0['start_time'].apply(sec_from_mid)
read_0['age'] = read_0['member_birth_year'].apply(get_age)
```

```python
plt.figure(1)
plt.subplot(211)
n,bins,pathces = plt.hist(read_0.groupby('weekday').get_group(True)['sec_from_mid'], bins = 200)
plt.title('Weekday')

plt.subplot(212)
n,bins,pathces = plt.hist(read_0.groupby('weekday').get_group(False)['sec_from_mid'], bins = 200)
plt.title('Weekend')
lab = plt.xlabel('Seconds from Midnight')

plt.tight_layout()
```

#### Here is a breakdown of frequency of trip start time for weekday/weekend and customer/subscriber.  There are several observations.  (Note the histogram bars are not stacked.)

#### First, weekday subscribers dominate the market by a factor of 10.  In addition, these trips follow a very clear commuter pattern.  We conclude the SF bikeshare is primarily a commuter system.

#### Second, weekend trips are evenly split between subscribers and customers.  In other words, most subscribers simply don't use the service on the weekend.  In general, all weekend trips follow a leisure pattern.  Rentals increase throughout the daytime until the afternoon, then decrease.  I hypothesize these are people running errands and tourists.

#### Third, note the heavy tale weekend trips around midnight.  I hypothesize these are people going out for nightlife events.

Specific questions would be: are the stations located well for all use cases?  Do all cities bikeshares follow a similar pattern, or do some cities have different market segmentation?

#### I will add a label to the data to classify the station based on city.  I use a k-means algorithm.  This is a trivial example because the cities are so well spaced.  However, it shows the use of clustering stations and will be applied in less obvious cases to find stations clusters in other cities.

```python
from sklearn.cluster import KMeans

model = KMeans(3)
result = model.fit(stations_df.iloc[:,1:].as_matrix())

result.labels_

stations_df['city'] = result.labels_

folium_map = folium.Map(location=[37.576435, -122.226244],
                        zoom_start=10,
                        tiles="CartoDB dark_matter")

for index, row in stations_df.iterrows():
    city = row['city']
    pos = (row['start_station_latitude'],row['start_station_longitude'])
    if city == 0:
        color = 'blue'
    elif city == 1:
        color = 'red'
    else:
        color = 'green'
       
    marker = folium.CircleMarker(location=pos,color= color)
    marker.add_to(folium_map)

folium_map
```

#### An interesting analysis is to break the network by user type.  In particular, we see that commuters are evenly distributed in the morning, but do not start trips from Berkley in the evening.  We conclude that there is a significant portion of people commuting from Oakland to SF.
